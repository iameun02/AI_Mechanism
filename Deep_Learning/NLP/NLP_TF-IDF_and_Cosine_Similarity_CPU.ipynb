# TF-IDF & Consine Similarity

### <b>TF-IDF</b>
 - 개별문장에 자주 나타나는 단어에 높은 가중치를 부여하고
   모든 문장에 전반적으로 자주 나타나는 단어에는 패널티를 부여

  - tf score : 특정문장에 등장한 특정단어의 등장횟수
  - idf score : log(전체문장의수/ 해당단어를 포함하는 문장의수)
  - tf-idf score :  tf score * idf score
<br><br>


----------------------------------------------------------------------
### <b>Similarity (유사도)</b>

1.  Euclidean Distance Similarity
    - 벡터간의 거리를 계산하여 유사도를 측정
    - 거리가 짧을수록 유사도가 높음

2. cosine 
   - 두 벡터간의 사잇각을 계산하여 유사도를 측정
   - 사잇각이 작을수록 유사도가 높음
   - 벡터의 크기가 아닌 방향성 기반

# II. cos_similarity( )

* 두 벡터의 크기와 상관없이, 상호 방향성이 얼마나 유사한지에 기반
 - 두 벡터의 사잇각을 계산하여 유사도 측정
<br>
<br>
* 두 벡터의 내적을 총 벡터 크기로 정규화(L2 Norm)
 - dot_product: 두 벡터의 내적
 - l2_norm: 총 벡터 크기의 합

```python
import numpy as np

def cos_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    l2_norm = (np.sqrt(sum(np.square(v1))) * np.sqrt(sum(np.square(v2))))
    similarity = dot_product / l2_norm     
    
    return similarity
```

> ## 1) 'vec1', 'vec2' 코사인 유사도

```python
similarity_simple = cos_similarity(vec1, vec2)

print('vec1, vec2 코사인 유사도: {0:.5f}'.format(similarity_simple))
```

> ## 2) 'vec1', 'vec3' 코사인 유사도

```python
similarity_simple = cos_similarity(vec1, vec3)

print('vec1, vec3 코사인 유사도: {0:.5f}'.format(similarity_simple))
```

> ## 3) 'vec2', 'vec3' 코사인 유사도

```python
similarity_simple = cos_similarity(vec2, vec3)

print('vec2, vec3 코사인 유사도: {0:.5f}'.format(similarity_simple))
```

# IV. Topic Problem

- 단어가 가지고 있는 의미를 분별하지 못함으로 인해 <br>
  유사도가 왜곡되는 이슈발생 ~> embedding 방식 발전
- embedding의 word2vec 은 '학습'기반의 vectorization을 수행

> ## 1) 문장 지정

```python
sent_list = ['I eat an apple',
             'Koo have fruit',
             'I sell an apple']
```

> ## 2) 벡터 변환

```python
print(feature_vec.toarray())
```

> ## 3) 문장1 vs. 문장2

```python
print(cosine_similarity(feature_vec[0], feature_vec[1]))
```

> ## 4) 문자1 vs. 문장3

```python
print(cosine_similarity(feature_vec[0], feature_vec[2]))
```

# V. Word2Vec


* 벡터 공간상에 비슷한 위치에 있다면 유사도(비슷한 의미)가 높음
* 문장단위가 아니라 단어 단위로 벡터화



> ## 1) Load Pretrained Word2Vec

```python
import tensorflow_hub as hub

embed = hub.load('https://tfhub.dev/google/Wiki-words-250/2')
```

> ## 2) 'words' List 정의

```python
words = ['apple', 'eat', 'fruit', 'have', 'sell']
```

> ## 3) Word2Vec Embedding

```python
embeddings = embed(words)
```

```python
embeddings #각행마다 한 단어를 표현 : 각 단어를 250개의 열로 표현
```

> ## 4) Cosine Similarity

```python
for i in range(len(words)):
    for j in range(i,len(words)):
        print("(",words[i], ",", words[j],")", cos_similarity(embeddings[i], embeddings[j]))
```
